```python
from dotenv import load_dotenv
import os
import requests
import pandas as pd

load_dotenv()
```

```python
API_KEY = os.getenv("API_KEY")
```

```python
# request list of members from stapi: cms.gdscmpstme.com/api/members using the api key
response = requests.get(
    "https://cms.gdscmpstme.com/api/members?populate=image&pagination[limit]=100",
    headers={"Authorization": f"Bearer {API_KEY}"}
)

members = response.json()
```

```python
members['data'][0]
```

```python
i = 0

for member in members['data']:
    # print the name of each member
    print(member["attributes"]["name"])
    print(member["attributes"]["designation"])
    print(member["attributes"]["year"])
    if member["attributes"]['name'] not in ['Pankaj Mirchandani', 'Vedant Maheshwari', 'Smartana Sonawane', 'Hisham Sopariwala']:
        print(member["attributes"]['image']['data']['attributes']['formats']['medium']['url'])
    print()

    i += 1
    if i == 5:
        break

print(f'{i} members found.')
```

```python
# update all members to have year = 2022
for member in members['data']:
    member_id = member["id"]

    response = requests.put(
        f"https://cms.gdscmpstme.com/api/members/{member_id}",
        headers={"Authorization": f"Bearer {API_KEY}"},
        json={"data": {
            "year": 2022
        }}
    )

    print(f'Updating {member_id}: {member["attributes"]["name"]}...')
```

```python
# convert members.csv to json
data = pd.read_csv('members.csv')

# data.columns = ['name', 'department', 'designation']
# rename columns
data.rename(columns={'Full Name': 'name', 'Department': 'department', 'Position': 'designation'}, inplace=True)

# add column 'year' to the dataframe
data['year'] = 2023

# add column 'image' to the dataframe
data['image'] = None

for i in range(len(data)):
    for image in images:
        if data.iloc[i]['name'] in image:
            img = image
            break

    data.at[i, 'image'] = f'./data/images/{img}'

data.to_json('members.json', orient='records')
```

```python
# upload data of all members to cms.gdscmpstme.com
members = json.load(open('members.json'))

for member in members:
    response = requests.post(
        "https://cms.gdscmpstme.com/api/upload",
        headers={"Authorization": f"Bearer {API_KEY}"},
        files={"files": open(member['image'], "rb")}
    )

    image = response.json()

    response = requests.post(
        "https://cms.gdscmpstme.com/api/members",
        headers={"Authorization": f"Bearer {API_KEY}"},
        json={"data": {
            "name": member['name'],
            "department": member['department'],
            "designation": member['designation'],
            "year": member['year'],
            "image": image[0]['id']
        }}
    )

    print("Created member: ", response.json()['data']['attributes']['name'])
```
